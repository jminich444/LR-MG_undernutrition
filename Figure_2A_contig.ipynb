```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
```

```python
#read in table used to make figure
contig = pd.read_csv("Supp_tables/Table_S7_Top100Contigs_Compare.csv", index_col=0)

```

```python
# Assuming your DataFrame is named df
sns.set(style="whitegrid")
approach_color= {
    'ONT': 'blue',
    'PacBio': 'magenta',
    'Tell-Seq_SLR': 'darkorange',
    'Short-read_Only': 'orange'
}

# Create the lineplot
plt.figure(figsize=(6, 6))
sns.lineplot(
    data=contig,
    x='Contig_number',
    y='Contig_Length_bp',
    hue='Approach',
    style='Assembler',
    markers=True,
    dashes=False,
    palette=approach_color
)
plt.gca().yaxis.set_major_formatter(ScalarFormatter())  # Keep y-axis in full numbers
plt.ticklabel_format(style='plain', axis='y')           # Disable scientific notation
plt.yscale('log')
plt.xlabel('Contig Number')
plt.ylabel('Contig Length (bp)')
plt.ylim(5e4, 1e7)
plt.tight_layout()
plt.savefig("Figure_2F_contig_plot.tiff", dpi=600, bbox_inches='tight')
```
