```python
#Pacbio Revio summary run statistics

df=pd.DataFrame(seqyield_lig)

#prints only the values within the column <platform> which are Pacbio Revio
#print(df.loc[df['platform'] == 'Pacbio Revio'])
pb=df.loc[df['platform'] == 'Pacbio Revio']
ont=df.loc[df['platform'] == 'ONT Promethion']

df.head()


```

```python
#saves the instrument reported values 'includes the reads that don't get assigned to barcodes'
### Pacbio results
pbIn=pb.loc[pb['data']=='instrument reported']
pbIn.describe()

#results
  #reads_M: mean, stdev (5.25, 1.48, n=14)
  #Yield_Gb: mean, stdev (42.24, 13.57, n=14)

pbAct=pb.loc[pb['data']=='post-processing']
pbAct.describe()

#results
  #reads_M: mean, stdev (4.94, 1.37, n=14)
  #Yield_Gb: mean, stdev (39.28, 12.42, n=14)
```

```python

#### ONT RESULTS
ontIn=ont.loc[ont['data']=='instrument reported']
ontIn.describe()
#results
  #reads_M: mean, stdev (20.37, 14.27, n=15)
  #Yield_Gb: mean, stdev (78.11, 46.05, n=15)

ontAct=ont.loc[ont['data']=='post-processing']
ontAct.describe()
#results
  #reads_M: mean, stdev (15.64, 10.51, n=15)
  #Yield_Gb: mean, stdev (56.85, 32.79, n=15)
```

```python
#statistical comparison Yield Gb
ont_gb=ontAct['Yield_Gb']
ont_reads=ontAct['reads_M']
pb_gb=pbAct['Yield_Gb']
pb_reads=pbAct['reads_M']

i_ont_gb=ontIn['Yield_Gb']
i_ont_reads=ontIn['reads_M']
i_pb_gb=pbIn['Yield_Gb']
i_pb_reads=pbIn['reads_M']

#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mannwhitneyu.html

##Gb differences
#Raw off-instrument values show that ONT yields are higher than PB for Gb and reads
#But when you take into account only useable data (>Q10 for ONT and >Q20 for Pacbio along with only reads which were successfully ligated with a barcodes, there are no differences in Gb yield)

## instrument values

#statistical test for yields Gb across platforms based on instrument values
stats.mannwhitneyu(i_ont_gb,i_pb_gb, use_continuity=True, alternative='two-sided')
#MannwhitneyuResult(statistic=165.0, pvalue=0.009409864199244939)

#statistical test for yields million reads across platforms (instrument yields)
stats.mannwhitneyu(i_ont_reads,i_pb_reads, use_continuity=True, alternative='two-sided')
#MannwhitneyuResult(statistic=172.0, pvalue=0.003700134957614701)

## actual post processing values

stats.mannwhitneyu(ont_gb,pb_gb, use_continuity=True, alternative='two-sided')
#testing if differences in yield (Gb) for actual post processing values
#MannwhitneyuResult(statistic=144.5, pvalue=0.08869742095268623)


stats.mannwhitneyu(ont_reads,pb_reads, use_continuity=True, alternative='two-sided')
#statistical test for yields million reads across platforms (actual yields)
#MannwhitneyuResult(statistic=165.0, pvalue=0.009409864199244939)
```

```python
# Useable data for ONT

ontAct.describe()
#useable data Gb (71.45%, 6.94) n=15
#useable reads (78.16%, 4.67)

pbAct.describe()
#useable data Gb (93.2, 2.08, n=14)
#useable data reads M (94.3, 1.51, n=14)

```

```python
## Stats on useable data
r_ont_gb=ontAct['per_retain_Gb']
r_ont_reads=ontAct['per_retain_reads']
r_pb_gb=pbAct['per_retain_Gb']
r_pb_reads=pbAct['per_retain_reads']

#significantly higher
stats.mannwhitneyu(r_ont_gb,r_pb_gb, use_continuity=True, alternative='two-sided')
#MannwhitneyuResult(statistic=0.0, pvalue=5.097028889963147e-06)

#significantly higher
stats.mannwhitneyu(r_ont_reads,r_pb_reads, use_continuity=True, alternative='two-sided')
MannwhitneyuResult(statistic=0.0, pvalue=5.097028889963147e-06)

```
