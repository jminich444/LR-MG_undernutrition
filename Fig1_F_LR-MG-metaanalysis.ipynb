```python
# meta-analysis LR-MG resource
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

sns.set(style="whitegrid")
```

```python
resource=pd.read_csv('../Table_S6_LRMG_meta.csv')
```

```python
res1=resource.loc[resource['include_']==1]
```

```python
fig, ax=plt.subplots(figsize=(4,4))
fig1=sns.scatterplot(data=res1, x="samples", y="total_Gbp", hue='technology', palette=['blue', 'magenta'], style="longitudinal", size="ONT_vs_PB", sizes=(100, 30))
ax.set(xlabel='Total samples in study', ylabel='Total Gbp in study')
plt.xticks(np.arange(0, 210, step=20))
plt.yticks(np.arange(0, 1300, step=100))
plt.savefig("../Figures/Fig1E_meta_LRMG.png", dpi=600, bbox_inches="tight")
```
